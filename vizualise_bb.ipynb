```python
import matplotlib.pyplot as plt

# Given coordinates
corners = [
    {"x": 526.157122891623, "y": 250.9300197470226},
    {"x": 490.66192424384093, "y": 258.5052145803907},
    {"x": 513.6039428820416, "y": 328.1970070473774},
    {"x": 551.479917048882, "y": 317.15886600446953}
]
center = {"x": 520.4757267665968, "y": 288.69777684481505}

# Extract x and y coordinates of corners
corner_x = [coord["x"] for coord in corners]
corner_y = [coord["y"] for coord in corners]

# Extract center coordinates
center_x = center["x"]
center_y = center["y"]

# Plot the corners and center
plt.figure(figsize=(6, 6))
plt.plot(corner_x + [corner_x[0]], corner_y + [corner_y[0]], marker='o', label='Corners')
plt.plot(center_x, center_y, marker='o', color='red', label='Center')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Corners and Center Plot')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import Polygon
from PIL import Image
def plot_bbox(image_path, annotation_path):
    # Load annotation JSON
    with open(annotation_path, 'r') as f:
        annotations = json.load(f)
    
    # Open the image
    image = Image.open(image_path)

    # Plot the image
    plt.figure(figsize=(10, 10))
    plt.imshow(image)

    # Plot bounding boxes
    for instance in annotations['instances']:
        corners = instance['corners']
        x_values = [corner['x'] for corner in corners]
        y_values = [corner['y'] for corner in corners]
        polygon = Polygon(xy=list(zip(x_values, y_values)), closed=True, fill=None, edgecolor='r')
        plt.gca().add_patch(polygon)
        
        label = "defect" if instance['defected_module'] else "good"
        plt.text(x_values[0], y_values[0], label, color='white', fontsize=8, bbox=dict(facecolor='red', alpha=0.5))

    plt.axis('off')
    plt.show()
```

```python
image_path = r"C:\Users\vrind\Downloads\segmentationnproject\datasets\Photovoltaic\images\20180630_154201.jpg"
annotations_path = r"C:\Users\vrind\Downloads\segmentationnproject\datasets\Photovoltaic\annotations\20180630_154201.json"
plot_bbox(image_path, annotations_path)
```
